```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor
```

# Metodo Random Forest

```python
df = pd.read_csv("./train.csv")
df['agregados']= df['escuelascercanas']+df['gimnasio']+df['centroscomercialescercanos']+df['piscina']+df['usosmultiples']
train = df.drop(['id','idzona','lat','direccion','ciudad','provincia','descripcion','titulo','lng', 'escuelascercanas','gimnasio','centroscomercialescercanos','piscina','usosmultiples'],1)
train['fecha'] = pd.to_datetime(train['fecha']).dt.year
train.fillna(0, inplace=True)
train
```

```python
train['tipodepropiedad'].value_counts()

def f (tipo):
    
    if tipo == "Apartamento": 
        return 1

    if tipo == "Bodega comercial": 
        return 2

    if tipo == "Casa en condominio": 
        return 3

    if tipo == "Casa uso de suelo": 
        return 4

    if tipo == "Departamento Compartido": 
        return 5

    if tipo == "Duplex": 
        return 6

    if tipo == "Edificio": 
        return 7

    if tipo == "Garage": 
        return 8

    if tipo == "Hospedaje": 
        return 9

    if tipo == "Huerta": 
        return 10

    if tipo == "Inmuebles productivos urbanos": 
        return 11

    if tipo == "Local Comercial": 
        return 12
    
    if tipo == "Local en centro comercial": 
        return 13

    if tipo == "Lote": 
        return 14
    
    if tipo == "Nave industrial": 
        return 15

    if tipo == "Oficina comercial": 
        return 16
    
    if tipo == "Otros": 
        return 17

    if tipo == "Quinta Vacacional": 
        return 18
    
    if tipo == "Rancho": 
        return 19
    
    if tipo == "Terreno": 
        return 20
    
    if tipo == "Terreno industrial": 
        return 21
    
    if tipo == "Villa": 
        return 22
    
    else: return 0

train['tipodepropiedad'] = train['tipodepropiedad'].apply(f)
```

```python
X = train.drop(['precio'],1) 
y = train.iloc[:, 8].values 
y
```

```python
from sklearn.model_selection import train_test_split
#Separo los datos de "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
from sklearn.model_selection import GridSearchCV# Create the parameter grid based on the results of random search 
param_grid = {
    'bootstrap': [True,False],
    'max_depth': [42, 30],
    'min_samples_leaf': [4],
    'min_samples_split': [8, 10],
    'n_estimators': [74, 84, 100],
    'random_state':[42]
}# Create a based model
rf = RandomForestRegressor()# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, cv = 3, verbose = 2)
print(grid_search)
```

```python
# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```
