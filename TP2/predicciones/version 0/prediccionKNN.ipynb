```python
import numpy as np
import pandas as pd

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
from sklearn.metrics import mean_absolute_error
from sklearn.neighbors import KNeighborsRegressor
```

```python
def valorizar_tipo_propiedad (tipo):
    
    if tipo == "Garage": 
        return 1

    if tipo == "Lote": 
        return 2

    if tipo == "Duplex": 
        return 3

    if tipo == "Terreno": 
        return 4

    if tipo == "Local en centro comercial": 
        return 5

    if tipo == "Local Comercial": 
        return 6

    if tipo == "Casa": 
        return 7

    if tipo == "Terreno comercial": 
        return 8

    if tipo == "Departamento Compartido": 
        return 9

    if tipo == "Bodega comercial": 
        return 10

    if tipo == "Apartamento": 
        return 11

    if tipo == "Villa": 
        return 12
    
    if tipo == "Otros": 
        return 13

    if tipo == "Casa en condominio": 
        return 14
    
    if tipo == "Hospedaje": 
        return 15

    if tipo == "Oficina comercial": 
        return 16
    
    if tipo == "Quinta Vacacional": 
        return 17

    if tipo == "Huerta": 
        return 18
    
    if tipo == "Terreno industrial": 
        return 19
    
    if tipo == "Casa uso de suelo": 
        return 20
    
    if tipo == "Nave industrial": 
        return 21
    
    if tipo == "Inmuebles productivos urbanos": 
        return 22

    if tipo == "Rancho": 
        return 23

    if tipo == "Edificio": 
        return 24
    
    else: return 0
```

#### Modelo 1: KNN

```python
trainB = train.drop(['id','descripcion', 'titulo', 'direccion', 'ciudad', 'fecha'], axis = 1)

#Adiciono servicios
trainB['servicios'] = trainB['piscina'] + trainB['gimnasio'] + trainB['usosmultiples']

#Adiciono cercanias
trainB['ubicacion'] = trainB['escuelascercanas'] + trainB['centroscomercialescercanos']

#Elimino los residuos
trainB = trainB.drop(['piscina', 'gimnasio','usosmultiples','escuelascercanas', 
                      'centroscomercialescercanos'], axis = 1)

#Valorizo tipo de propiedad del anuncio
trainB['tipodepropiedad'] = trainB['tipodepropiedad'].apply(valorizar_tipo_propiedad)

#Valorizo provincia del anuncio
trainB['provincia'] = trainB['provincia'].apply(valorizar_provincia)

trainB = trainB.dropna()

#Preparo Train
X = trainB.drop(['precio'], axis = 1)
Y = trainB['precio']

neigh = KNeighborsRegressor(n_neighbors = 2)

neigh.fit(X,Y)

testB = test.drop(['id','descripcion', 'titulo', 'direccion', 'ciudad', 'fecha'], axis = 1)

#Adiciono servicios
testB['servicios'] = testB['piscina'] + testB['gimnasio'] + testB['usosmultiples']

#Adiciono cercanias
testB['ubicacion'] = testB['escuelascercanas'] + testB['centroscomercialescercanos']

#Elimino los residuos
testB = testB.drop(['piscina', 'gimnasio','usosmultiples','escuelascercanas', 
                      'centroscomercialescercanos'], axis = 1)

#Valorizo tipo de propiedad del anuncio
testB['tipodepropiedad'] = testB['tipodepropiedad'].apply(valorizar_tipo_propiedad)

#Valorizo provincia del anuncio
testB['provincia'] = testB['provincia'].apply(valorizar_provincia)

testB = testB.fillna(testB.mean())


prediccion = neigh.predict(testB)

#prediccion = neigh.predict(X) #make prediction on test set
#errorMAE = mean_absolute_error(Y,prediccion) #calculate mae
#errorMAE

prediccion
```

```python
predicciones = test.loc[:, ['id','ciudad']]
predicciones.rename(columns = {'ciudad' : 'target'}, inplace = True)
predicciones['target'] = prediccion
predicciones = predicciones.set_index('id')
```
