```python
import pandas as pd
import matplotlib.pyplot as plt

estimadores = {
                'estimadores': [250, 300, 400, 458, 700, 900, 1100, 1700, 3500],
                'tiempo': [118.50774515500234, 198.24549748099525, 422.0792062029941, \
                           490.03200192899385, 474.65883604499686, 714.0491829480015, \
                           496.1875175340101, 1145.358375625001, 4244.4395449540025]
}

df = pd.DataFrame(estimadores)
```

```python
fig, ax1 = plt.subplots() # prepara un gráfico de matplotlib
df.plot("estimadores", "tiempo", kind="bar", ax=ax1, figsize=(7,7), color='blue')

# realiza ajustes al gráfico con matplotlib:
ax1.grid()
ax1.set_axisbelow(True)
ax1.set_xlabel("Cantidad de estimadores", fontsize=15)
ax1.set_ylabel("Tiempo de demora del fit (segundos)", fontsize=15)
ax1.tick_params(labelsize=10, pad=10, rotation=0)
fig.suptitle('Tiempo de demora de la ejecución del fit en relación a la cantidad de estimadores', fontsize=20)

#plt.savefig('/home/ailen-magali/Documentos/precio_por_categorias.png', bbox_inches = 'tight', pad_inches = 0.1)
```
